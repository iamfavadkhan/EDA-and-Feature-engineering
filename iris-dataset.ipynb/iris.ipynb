```python
#Load and Read the Data
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv")
```

```python
#Inspect the data
df = pd.DataFrame(data)
df.head()
```

```python
df.rename(columns={'sepal.length':'Sepal_length','sepal.width':'Sepal_width'},inplace=True)
df.columns
```

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

```python
print(df['variety'].unique())
```

```python
df.hist(figsize=(10,6), bins=15)
plt.show()
```

```python
sns.boxplot(data=df, orient="h")
plt.show()
```

```python
sns.countplot(x="variety", data=df)
plt.show()
```

```python
df.columns
```

```python
sns.scatterplot(x="Sepal_length", y="Sepal_width", hue="variety", data=df)
plt.show()
```

```python
sns.pairplot(df , hue="variety")
plt.show()
```

```python
corr = df.drop(columns="variety").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
for col in df.select_dtypes(include="number").columns:
    sns.boxplot(x=df[col])
    plt.show()
```

```python
df.to_csv("cleaned_iris.csv", index=False)
```
