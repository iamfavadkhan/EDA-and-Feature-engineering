#Step 1: Load and Explore the Dataset

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

data = pd.read_csv("delhi-air-pollution.csv")
df = pd.DataFrame(data) #It wil convert the dataset into DataFrame
df.head(10) #it will print 1st 10 row of the dataset

```

```python
df.info() #Will show the number columns and necessory information
df.describe() #Will count all the basic methods for the dataset 
```

```python
print(df[["PM2.5" , "PM10"]].fillna(0)) #This will fill the null values by 0
print(df.isnull().sum()) #This will show the total sum of null values
```

#Step 2: Data Cleaning

```python
df_filtered = df[(df['Year'] >= 2020) & (df['Year'] <= 2024)]
df_filtered
```

#Step 3:
Part A: Analysis part

```python
grouped = df.groupby(['Month' , 'Year'])[["PM2.5" , "PM10" , "NO2" , "AQI"]].mean().reset_index() #This will group the data into month and year and then count the mean of each group
#rouped.head(20)   #It print the first 20 rows 
grouped.sample(20) #It will print random 20 rows from the dataset
```

#Step 3:
Part B:Pollution Severity Categorization

```python
conditions = [
    (df['AQI'] <= 50),
    (df['AQI'] >= 50) & (df['AQI'] <= 100),
    (df['AQI'] >= 101) & (df['AQI'] <= 200),
    (df['AQI'] >= 201) & (df['AQI'] <= 300),
    (df['AQI'] >= 301) & (df['AQI'] <= 400),
    (df['AQI'] >= 401)
]

choices = [
    'Good',
    'Satisfactory',
    'Moderate',
    'Poor',
    'Very',
    'Severe'
]

df["aqi_categories"] = np.select(conditions , choices , default='Unknown') #it will categorize the whole dataset into above categories
df["aqi_categories"].sample(20)
```

Step 3:
Part C: Peak Pollution Periods

```python
top_5_polluted = df.sort_values(by='AQI', ascending=False).head(5)
top_5_polluted
```

```python
pm25_avg = df.groupby(['Month' , 'Year'])['PM2.5'].mean().reset_index()
max_pm25 = pm25_avg.loc[pm25_avg['PM2.5'].idxmax()]
max_pm25
```

#Step 4: Visualization

```python
# ðŸ“ˆ Line Plot
# Monthly average PM2.5 and PM10 trends (2020â€“2.024)

sns.lineplot(x='PM2.5' , y='PM10' , data = df , hue = 'aqi_categories')
plt.title("Monthly average PM2.5 and PM10 trends (2020â€“2.024)")
plt.figure(figsize=(12,8))
plt.show()
```

```python
# ðŸ“Š Bar Chart
# Average AQI per year.
plt.figure(figsize=(8,6))
avg_years = df.groupby('Year')['AQI'].mean().reset_index()
sns.barplot( data = avg_years , x='AQI' , y='Year')
plt.title("Average AQI Per Year")
plt.xlabel("Average AQI")
plt.ylabel("Year")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# import matplotlib.pyplot as plt

# Group PM2.5 by year
years = sorted(df['Year'].unique())
data_by_year = [df[df['Year'] == year]['PM2.5'].dropna() for year in years]

plt.figure(figsize=(10, 6))
plt.boxplot(data_by_year, labels=years)
plt.title('PM2.5 Spread by Year')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration')
plt.grid(True)
plt.show()
```

```python
pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'Ozone', 'AQI']
corr_matrix = df[pollutants].corr()

plt.figure(figsize=(14, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap Between Pollutants')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df['AQI'].dropna(), shade=True, color='red')
plt.title('AQI Distribution (KDE)')
plt.xlabel('Air Quality Index')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
